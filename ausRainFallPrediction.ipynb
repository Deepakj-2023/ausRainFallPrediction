```python
# outliers treatments
# only for the continues variable
```

```python
def wisker(col):
    q1,q3 =np.percentile(col,[25,75])
    iqr = q3-q1
    uw = q3+1.5*iqr
    lw =q1-1.5*iqr
    return lw,uw
```

```python
wisker(df['Humidity3pm'])
```

```python
for i in ['Humidity3pm','Pressure9am','Pressure3pm']:
    lw,uw = wisker(df[i])
    df[i] = np.where(df[i]<lw,lw,df[i])
    df[i] = np.where(df[i]>uw,uw,df[i])
    
```
